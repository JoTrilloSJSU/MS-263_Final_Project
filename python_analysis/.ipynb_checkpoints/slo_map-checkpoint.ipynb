## Cartopy tutorial

This tutorial uses the Cartopy package to make maps.

Geopandas is newer package for plotting and analyzing geospatial data. It is not covered in this tutorial, but it is worth exploring: https://geopandas.readthedocs.io

Xarray is used to load chlorophyll-a and bathymetry data in NetCDF format

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy.crs as ccrs
import cmocean.cm as cmo
```

### Plotting satellite chlorophyll-a

Data obtained from the NASA Ocean Color website
https://oceancolor.gsfc.nasa.gov/

Go to Data --> Level 3 Browser to search gridded images

```python
chl_nc = 'nasa_ocean_color/A20220812022088.L3m_8D_CHL_chlor_a_9km.nc'
```

```python
ds = xr.open_dataset(chl_nc)
ds
```

### Map projections

```python
plt.figure()

plt.pcolormesh(ds['lon'], ds['lat'], np.log10(ds['chlor_a']), vmin = -2, vmax = 2, cmap = 'turbo')
plt.colorbar()
plt.show()
```

```python
plt.figure()

ax = plt.axes(projection = ccrs.Orthographic(central_longitude = 90))
plt.pcolormesh(ds['lon'], ds['lat'], np.log10(ds['chlor_a']),
               vmin = -2, vmax = 2, cmap = 'turbo', transform = ccrs.PlateCarree())
#plt.colorbar()
ax.coastlines()
plt.show()
```

### Monterey Bay coastline

```python
# map extent for Monterey Bay
extent = [-122.3, -121.6, 36.5, 37]
```

```python
plt.figure()
ax = plt.axes(projection=ccrs.Mercator())

ax.coastlines()

gl = ax.gridlines(crs = ccrs.PlateCarree(), draw_labels = True)

ax.set_extent(extent, crs = ccrs.PlateCarree() )
```

```python
import cartopy.io.img_tiles as cimgt

request = cimgt.GoogleTiles(style = 'satellite')

plt.figure()
ax = plt.axes(projection=ccrs.Mercator())

ax.coastlines()
ax.add_image(request, 12) #lower number bigger pixels, bigger number higher resolution (max 18?)

gl = ax.gridlines(crs = ccrs.PlateCarree(), draw_labels = True)

ax.set_extent(extent, crs = ccrs.PlateCarree() )
```

### Monterey Bay bathymetry

The NOAA Coastal Relief Model is a gridded bathymetry product for US waters
https://www.ncei.noaa.gov/products/coastal-relief-model

```python
#bathy_nc = 'https://www.ngdc.noaa.gov/thredds/dodsC/crm/crm_vol6.nc'
bathy_nc = 'coastal_relief_model/crm_vol6_subset.nc'
```

```python
ds = xr.open_dataset(bathy_nc)
ds
```

```python
masked_depth = np.ma.masked_where(ds['z'] > 0, ds['z']) # conditions for masking, what variable you want to mask
```

```python
request = cimgt.GoogleTiles(style = 'satellite')

plt.figure()
ax = plt.axes(projection=ccrs.Mercator())

ax.coastlines()
ax.add_image(request, 12) #lower number bigger pixels, bigger number higher resolution (max 18?)

ax.pcolormesh(ds['x'], ds['y'], masked_depth, transform = ccrs.PlateCarree(), cmap = cmo.deep_r)
ax.contour(ds['x'], ds['y'], -1*masked_depth, levels = np.arange(0, 3800, 200), colors = 'black', 
           transform = ccrs.PlateCarree(), linewidths = 0.5)

gl = ax.gridlines(crs = ccrs.PlateCarree(), draw_labels = True)

ax.set_extent(extent, crs = ccrs.PlateCarree() )
```

### Shapefiles

```python
shp_file = 'Elkhorn_Slough_GIS/WETLAND/wetland.shp'
```

```python
from cartopy.io import shapereader
```

```python
wetland = shapereader.Reader(shp_file)
```

```python
plt.figure()
ax = plt.axes(projection = ccrs.Mercator())

for geometry in wetland.geometries():
    ax.add_geometries(geometry, ccrs.UTM(zone = 10), facecolor = 'white', edgecolor = 'black')


ax.set_extent((-121.8, -121.72, 36.79, 36.88), crs = ccrs.PlateCarree())
ax.gridlines(crs = ccrs.PlateCarree(), draw_labels = True)
```
